```python
import json,nltk,numpy as np
from nltk.stem import WordNetLemmatizer
```

```python
lem = WordNetLemmatizer()
```

```python
with open('../data/intents.json', encoding='utf-8') as f:
    data = json.load(f)
```

```python
sentences,labels = [],[]
for intent in data['intents']:
    for pattern in intent['patterns']:
        sentences.append(pattern.lower())
        labels.append(intent['tag'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vector = TfidfVectorizer()
x = vector.fit_transform(sentences)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y = encoder.fit_transform(labels)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([
    Dense(128,input_shape=(x.shape[1],),activation='relu'),
    Dense(64,activation='relu'),
    Dense(len(set(y)),activation='softmax')
])
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(x.toarray(), y, epochs=200, batch_size=8)
```

```python
import pickle
model.save('chatbot_model.h5')
pickle.dump(vector, open('../model/vectorizer.pkl', 'wb'))
pickle.dump(encoder, open('../model/label_encoder.pkl', 'wb'))
```
